# Generator

```python
import sys, os
import chainer
from chainer import serializers
import numpy as np
from gan import Generator
from matplotlib import pyplot as plt
z_dim = 100
```

```python
def draw_diget(data, row, col, n):
    size = 28
    plt.subplot(row, col, n)
    plt.imshow(data)
    plt.gray()
```

```python
model = Generator(z_dim)
serializers.load_npz("result\generator_iter_4282000.npz", model)
```

```python
num = 4
z = np.random.random((num*num, z_dim)).astype(np.float32)
z = chainer.Variable(z)
```

```python
x = model(z)
x = x.data
print(x.shape)
```

```python
plt.figure(figsize=(20, 20))
total = 0
for i in np.arange(num):
    for j in np.arange(num):
        draw_diget(x[i + num*j, 0], num, num, total+1)
        total += 1
plt.show()
```
